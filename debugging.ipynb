```python
# Get latest check-out from records.json
import os
import json
import pandas as pd
from datetime import datetime

# Define path of parameters
records_json = os.path.join(os.getcwd(), 'data', 'records.json')

def get_latest_record(records_json):

    with open(records_json) as f:
        records = json.load(f)

    last_cum_dict = records[-1]['cum']

    last_cum_df = pd.DataFrame(last_cum_dict)

    return last_cum_df

clean_df = get_latest_record(records_json)

clean_df
```

```python
def filter_df_on_checkout_date(clean_df, today=None):

    # Get today's date
    if not today:
        today = datetime.today().date()
        today = today.strftime('%d-%m-%Y')

    # Filter the DataFrame for check-in dates after today
    filtered_df = clean_df[pd.to_datetime(clean_df['Check-out date'], dayfirst=True) == today]

    return filtered_df

filtered_df = filter_df_on_checkout_date(clean_df, '10-02-2024')

filtered_df
```
